```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
def utility(x):
    return(np.exp(x ** gamma))
```

```python
pi = np.random.uniform(0,1,20)
pi = np.sort(pi)
```

```python
print('The highest three probability of losses are {}'.format(pi[-3:]))
```

```python
y = 2
c = 1.5
Q = 5
D = 0.01
gamma = 0.4
```

```python
def supply(Q):
    return(np.mean(pi[-Q:]) * c)
```

```python
def demand(D):
    return(np.sum(utility(y - D) > pi * utility(y - c) + (1 - pi) * utility(y)))
```

```python
plt.figure()
plt.plot([demand(i) for i in np.arange(0, 1.9, 0.02)],
         np.arange(0, 1.9, 0.02), 'r', label='insurance demand')
plt.plot(range(1, 21), [supply(j)
         for j in range(1, 21)], 'g', label='insurance supply')
plt.ylabel("Average Cost")
plt.xlabel("Number of People")
plt.legend()
plt.savefig('images/adverse_selection.png')
plt.show()
```
